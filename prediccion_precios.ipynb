**Imputamos por la moda las siguientes variables porque tienen muy pocos valores nulos (menores al 1% en todos los casos), por lo que imputar con la moda no intoduce sesgos importantes.**

```python
# Lista de variables a imputar con la moda
vars_none = ["MSZoning", "Exterior1st", "Exterior2nd", "Electrical", "KitchenQual", "SaleType"]

# Calcular la moda de cada variable en train
modes = train[vars_none].mode().iloc[0]

# Aplicar la imputación con la moda en train y test
for col in vars_none:
    train[col].fillna(modes[col], inplace=True)
    test[col].fillna(modes[col], inplace=True)
```

**Imputación de la variable 'GarageYrBlt'**

```python
# Calcular la correlación entre YearBuilt y GarageYrBlt
correlacion_garage_year = train["YearBuilt"].corr(train["GarageYrBlt"])

# Mostrar resultado
print(f"Correlación entre YearBuilt y GarageYrBlt: {correlacion_garage_year:.2f}")
```

**Ahora utilizamos el método de los cuartiles (IQR) para ver cuantos outliers tiene cada variable**

```python
 # Calcular outliers usando el método IQR
Q1 = train_selected.quantile(0.25)
Q3 = train_selected.quantile(0.75)
IQR = Q3 - Q1

# Filtrar outliers
outliers = ((train_selected < (Q1 - 1.5 * IQR)) | (train_selected > (Q3 + 1.5 * IQR))).sum()

# Ordenar los outliers de mayor a menor
outliers_sorted = outliers.sort_values(ascending=False)

# Mostrar los outliers ordenados
print("Número de outliers por variable:\n", outliers_sorted)
```

# Transformaciones

**Aplicamos transformación logaritmica a variables con una distribución sesgada (outliers altos)**

```python
vars_log = ["EnclosedPorch", "BsmtFinSF2", "LotFrontage", "ScreenPorch", "MasVnrArea", "OpenPorchSF",
            "LotArea", "TotalBsmtSF", "MiscVal","WoodDeckSF","SalePrice"]

# Aplicar transformación logarítmica en train
train[vars_log] = np.log1p(train[vars_log])

# Excluir 'SalePrice' de la lista para el conjunto de prueba
vars_log_test = [var for var in vars_log if var != "SalePrice"]
# Aplicar la transformación en test sin 'SalePrice'
test[vars_log_test] = np.log1p(test[vars_log_test])
```

**Eliminamos los valores que superaban el percentil 99 (outliers extremos) en algunas variables numéricas porque al analizar sus distribuciones vimos que estaban muy sesgadas y presentaban valores extremos.**

```python
vars_outliers = ["GrLivArea", "1stFlrSF", "2ndFlrSF", "GarageArea", "BsmtUnfSF"]

for col in vars_outliers:
    percentile_99 = train[col].quantile(0.99)
    train = train[train[col] < percentile_99]
```

**Aplicar Winsorización (capar valores al percentil 99) en las 4 siguientes variables porque al ser variables discretas no quería eliminarlas completamente como en el caso de los outliers continuos, pero sí corregir valores extremos que aparecen muy pocas veces y podrían distorsionar el entrenamiento del modelo. La Winsorización reemplaza esos valores extremos por un umbral (en este caso, el valor del percentil 99).**

```python
vars_winsor = ["BedroomAbvGr", "TotRmsAbvGrd", "GarageCars", "Fireplaces"]
for col in vars_winsor:
    percentile_99 = train[col].quantile(0.99)
    train[col] = np.where(train[col] > percentile_99, percentile_99, train[col])
```

**Aplicar Binarización (Tiene/No tiene) ya que a mayoría de las casas tienen 0-2 baños en el sótano. Puede ser útil marcar si tienen o no.**

```python
binary_vars = ["BsmtFullBath", "BsmtHalfBath", "FullBath", "HalfBath"]
for var in binary_vars:
    train[f"Has_{var}"] = (train[var] > 0).astype(int)
    test[f"Has_{var}"] = (test[var] > 0).astype(int)
```

# Selección de variables

Identificamos las variables numéricas que tienen más relación con la variable objetivo 'SalePrice', usando:
- Correlación Lineal
- F-Regression

He hecho esto para seleccionar solo las variables más informativas y relevantes para el modelo, reduciendo ruido y mejorando el rendimiento. Uso ambos métodos porque juntos dan una visiñon mas completa.

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import f_regression

# Separar variables predictoras y objetivo
X = train.drop("SalePrice", axis=1)
y = train["SalePrice"]

# Filtrar solo variables numéricas
X_numeric = X.select_dtypes(include=['number'])

# Calcular la correlación lineal entre cada variable y el target
correlation = X_numeric.corrwith(y).abs()

# Seleccionar variables con correlación alta (umbral > 0.5, ajustable)
selected_by_corr = correlation[correlation > 0.5].index.tolist()

# Opcional: Aplicar F-regression como verificación
f_values, p_values = f_regression(X_numeric.fillna(0), y)
f_scores = pd.Series(f_values, index=X_numeric.columns)

# Unir los criterios: alta correlación y alto F-score
selected_vars = list(set(selected_by_corr).intersection(set(f_scores[f_scores > 50].index)))

# Visualizar las variables seleccionadas
selected_vars.sort()
selected_vars
```

Estas son las variables que vamos a seleccionar para el modelo.

# Preparación del modelo

**División de X_train, X_test, y_train, y_test**


```python
from sklearn.model_selection import train_test_split

selected_features = [
    '1stFlrSF', 'FullBath', 'GarageArea', 'GarageCars',"ExterQual","KitchenQual","BsmtQual",
    'GarageYrBlt', 'GrLivArea', 'OverallQual', 'YearBuilt', 'YearRemodAdd',"Neighborhood",
]

X_train, X_test, y_train, y_test = train_test_split(
    train[selected_features],
    train["SalePrice"],
    random_state=0
)
```

**Pipeline: MinMaxScaler + SVR**

He aplicado MinMaxScaler para normalizar las variables numéricas antes de entrenar el modelo de SVR. MinMaxScaler transforma todas las variables al mismo rango [0,1], esto permite que todas las variables contribuyan por igual al modelo.

```python
from sklearn.svm import SVR
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler

# Pipeline: escalado + SVR
pipeline = Pipeline([
    ('scaler', MinMaxScaler()),
    ('svr', SVR())
])
```

**GridSearchCV con validación cruzada**

```python
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import mean_squared_error

# Grid de hiperparámetros
param_grid = {
    'svr__C': [1, 10, 100],
    'svr__epsilon': [0.01, 0.05, 0.1, 0.5],
    'svr__kernel': ['rbf'],
    'svr__gamma': ['scale', 'auto']
}

# Grid Search con validación cruzada
grid_search = GridSearchCV(
    estimator=pipeline,
    param_grid=param_grid,
    scoring='neg_mean_squared_error',
    cv=5,
    verbose=1,
    n_jobs=-1
)
```

```python
# Entrenar modelo
grid_search.fit(X_train, y_train)
```

```python
# Predicción y cálculo del RMSE
y_pred = grid_search.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
len(list(y_pred))
```

```python
# Resultados
print("Mejores hiperparámetros encontrados:", grid_search.best_params_)
print("RMSE en test:", rmse)
```

Este valor de RMSE en escala logaritmica indica un ajuste bastante bueno. Un RMSE log de 0.17 suele equivaler a un error relativo entre el 15% y 18% en los precios reales.
